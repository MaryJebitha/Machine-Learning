```python
import pandas as pd
```

```python
dataset =pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first =True)
dataset
```

```python
dataset.columns
```

```python
independent =dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
independent
```

```python
dependent =dataset[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(independent,dependent,test_size = 0.30, random_state =0)
```

```python
#model creation 
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='poisson', splitter='random')
regressor =regressor.fit(x_train,y_train)# in fit method only model is getting created and stored inside regressor variable
```

```python
# Decision Tree plot graph
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)# model creates is stored in reressor variable. for the created model we are plotting graph
plt.show()
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename ="finalized_model_DT.sav"
pickle.dump(regressor,open(filename,'wb'))
    
```
