```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=dataset['charges']
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
scy = StandardScaler()

X_train
```

 Reshape before scaling 
The reason to reshape y_train and y_test into 2D arrays before scaling is because of how scikit-learn’s StandardScaler is designed:

StandardScaler expects input in the form of a 2D array with shape (n_samples, n_features).
 target variable (y_train, y_test) is a 1D Series (shape (n_samples,)).
When I pass a 1D array directly, StandardScaler doesn’t know if I mean:
multiple samples with one feature each, or one sample with multiple features.
To avoid ambiguity, scikit-learn enforces the requirement that the input must be explicitly 2D.


The difference comes down to how scikit-learn interprets the shape of your data:

X_train and X_test are already 2D arrays (shape (n_samples, n_features)), because you selected multiple columns ([['age', 'bmi', 'children','sex_male', 'smoker_yes']]).
Example: if you have 1000 rows and 5 features, the shape is (1000, 5). That’s exactly what StandardScaler expects.

y_train and y_test, on the other hand, are 1D Series (shape (n_samples,)).
Example: (1000,) — just one column of values. StandardScaler doesn’t accept this directly, because it requires a 2D input where each column is treated as a separate feature.

So:

X_train → fine, already 2D.

y_train → needs reshaping into (n_samples, 1) so that StandardScaler knows it’s feature column."""

```python


#.reshape(-1, 1) converts your 1D array into a 2D array with one column (one feature).one -1 tells NumPy to infer the number of rows automatically.

y_train = scy.fit_transform(y_train.values.reshape(-1, 1))
y_test = scy.transform(y_test.values.reshape(-1, 1))
y_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
              'C':[10,100,1000,2000,3000],'gamma':['auto','scale']} 



grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1) 
   
# fitting the model for grid search 
grid.fit(X_train, y_train) 
 

```

```python
# print best parameter after tuning 
#print(grid.best_params_) 
re=grid.cv_results_

grid_prediction = grid.predict(X_test)

from sklearn.metrics import r2_score
r_score = r2_score(y_test,grid_prediction)

print("The R_score value for best parameter {}:".format(grid.best_params_),r_score)

```
